```python
def find_twitter_mentions(text):
    regex = "@(\w+)"
 
    # extracting the hashtags
    hashtag_list = re.findall(regex, text)
    return(hashtag_list)
    
```

```python
mentions['mentions'] = mentions['doc.text'].apply(find_twitter_mentions)
mentions[['doc.text','mentions']].head(20)
```
